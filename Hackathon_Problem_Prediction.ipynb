```python
import pandas as pd

train = pd.read_csv('Train.csv')
test = pd.read_csv('Test.csv')
```

```python
train.head(5)
```

#### Filling the missing value in street Type with mean

```python
train['Street Type'].fillna((train['Street Type'].mean()), inplace=True)
```

#### Filling the missing value in Actual Date of Completion with Estimated Date of Completion

```python
train['Actual Date of Completion1'] = train['Actual Date of Completion'].fillna(train['Estimated Date of Completion'])
```

#### Converting Date time Variable to datetime format

```python
train['Actual Date of Completion1'] =  pd.to_datetime(train['Actual Date of Completion1'], format='%d-%m-%Y %H:%M')
```

```python
train['Date of Creation'] =  pd.to_datetime(train['Date of Creation'], format='%d-%m-%Y %H:%M')
```

#### Created a new variable 'Time taken for resolution' 

```python
train['Time taken for resolution'] = train['Actual Date of Completion1'].sub(train['Date of Creation'], axis=0)
train.head()
```

```python
train['Time taken for resolution'] = train['Time taken for resolution'].dt.total_seconds() / 60
train['Time taken for resolution'].fillna((train['Time taken for resolution'].mean()), inplace=True)
```

```python
train.head()
```

```python
missing_data = pd.DataFrame({'total_missing': train.isnull().sum(), 'perc_missing': (train.isnull().sum()/train.shape[0])*100})
missing_data
```

#### Dropping the unnecessary feature that are not contributing much to target 

```python
X = train.drop(['Problem Category','L_Id','Date of Creation','Agent Category Assigned','Estimated Date of Completion','Actual Date of Completion','Actual Date of Completion1','A_1'],axis=1)
y = train['Problem Category']  # selecting the Purchase column to create labels
print(X.shape)
print(y.shape)
```

```python
X.head(5)
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaled_data=scaler.fit_transform(X)
```

```python
from numpy import loadtxt
from xgboost import XGBClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

```python
seed = 7
test_size = 0.33
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42)
```

```python
# fit model no training data
model = XGBClassifier()
model.fit(X_train, y_train)
```

```python
# make predictions for test data
y_pred = model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
# evaluate predictions
accuracy = accuracy_score(y_test, predictions)
print("Accuracy: %.2f%%" % (accuracy * 100.0))
```

```python
# evaluate predictions
from sklearn.metrics import f1_score
accuracy = 100*f1_score(y_test, predictions,average='macro')
```

```python
from sklearn.model_selection import GridSearchCV
param_grid={
   
    ' learning_rate':[1,0.5,0.1,0.01,0.001],
    'max_depth': [3,5,10,20],
    'n_estimators':[10,50,100,200]
    
}
```

```python
grid= GridSearchCV(XGBClassifier(objective='multi:softmax'),param_grid, verbose=3)
```

```python
grid.fit(X_train,y_train)
```

```python
grid.best_params_
```

```python
# Create new model using the same parameters
new_model=XGBClassifier(grid.best_params_)
new_model.fit(X_train, y_train)
```

```python
# make predictions for test data
y_pred = new_model.predict(X_test)
predictions = [round(value) for value in y_pred]
```

```python
# evaluate predictions
from sklearn.metrics import f1_score
accuracy = 100*f1_score(y_test, predictions,average='macro')
accuracy
```
